```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def removeLoop(head):
    if head is None or head.next is None:
        return head
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break
    if fast is None or fast.next is None:
        return head
    slow = head
    while slow.next != fast.next:
        slow = slow.next
        fast = fast.next
    fast.next = None
    return head


def printList(node):
    seen = set()
    while node and node not in seen:
        print(node.data, end=" ")
        seen.add(node)
        node = node.next

def createList():
    n = int(input("Enter the number of nodes: "))
    nodes = [None] * n
    for i in range(n):
        data = int(input(f"Enter data for node {i+1}: "))
        nodes[i] = Node(data)
        if i > 0:
            nodes[i-1].next = nodes[i]
    x = int(input("Enter an index to form a loop (0 for no loop): "))
    if x > 0 and x <= n:
        nodes[-1].next = nodes[x-1]
    return nodes[0]

def test():
    head = createList()
    print("The original linked list has a loop:", bool(detectLoop(head)))
    new_head = removeLoop(head)
    print("The new linked list has a loop:", bool(detectLoop(new_head)))
    print("The new linked list without loop is: ")
    printList(new_head)


def detectLoop(head):
    slow = head
    fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False

test()
```

```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def addOne(head):
    if head is None:
        return Node(1)
    curr = head
    last_not_nine = None
    while curr:
        if curr.data != 9:
            last_not_nine = curr
        curr = curr.next
    if last_not_nine is None:
        new_head = Node(1)
        new_head.next = head
        curr = head
        while curr:
            curr.data = 0
            curr = curr.next
        return new_head

    else:
        last_not_nine.data += 1
        curr = last_not_nine.next
        while curr:
            curr.data = 0
            curr = curr.next
        return head


def printList(node):
    while node:
        print(node.data, end="")
        node = node.next


def createList():
    n = int(input("Enter the number of nodes: "))
    nodes = [None] * n
    for i in range(n):
        data = int(input(f"Enter data for node {i+1}: "))
        nodes[i] = Node(data)
        if i > 0:
            nodes[i-1].next = nodes[i]
    return nodes[0]

def test():
    head = createList()
    print("The original linked list is: ")
    printList(head)
    print()
    new_head = addOne(head)
    print("The new linked list is: ")
    printList(new_head)


test()
```

```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.bottom = None

def merge(a, b):
    if a is None:
        return b
    if b is None:
        return a
    result = Node(0)
    temp = result
    while a and b:
        if a.data < b.data:
            temp.bottom = a
            a = a.bottom
        else:
            temp.bottom = b
            b = b.bottom
        temp = temp.bottom
    if a:
        temp.bottom = a
    if b:
        temp.bottom = b
    return result.bottom

def flatten(root):
    if root is None or root.next is None:
        return root

    return merge(root, flatten(root.next))

def printList(node):
    while node:
        print(node.data, end=" ")
        node = node.bottom

def createList():
    n = int(input("Enter the number of sub-linked-lists: "))
    heads = [None] * n
    for i in range(n):
        m = int(input(f"Enter the number of nodes in sub-linked-list {i+1}: "))
        head = None
        prev = None

        for j in range(m):
            data = int(input(f"Enter data for node {j+1} in sub-linked-list {i+1}: "))
            node = Node(data)
            if j == 0:
                head = node
                heads[i] = head
            else:
                prev.bottom = node
            prev = node
    return heads


def test():
    heads = createList()
    for i in range(len(heads) - 1):
        heads[i].next = heads[i+1]
    root = flatten(heads[0])
    print("The flattened linked list is: ")
    printList(root)


test()
```

```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.arb = None


def copyList(head):
    if head is None:
        return None
    curr = head
    while curr:
        new_node = Node(curr.data)
        new_node.next = curr.next
        curr.next = new_node
        curr = curr.next.next
    curr = head
    while curr:
        if curr.arb:
            curr.next.arb = curr.arb.next
        curr = curr.next.next
    new_head = head.next
    old_curr = head
    new_curr = new_head
    while old_curr:
        old_curr.next = old_curr.next.next
        old_curr = old_curr.next
        if new_curr.next:
            new_curr.next = new_curr.next.next
            new_curr = new_curr.next
    return new_head

def printList(node):
    seen = set()
    while node and node not in seen:
        print(f"({node.data}, {node.arb.data if node.arb else None})", end=" ")
        seen.add(node)
        node = node.next

def createList():
    N = int(input("Enter number of nodes: "))
    nodes = [None] * N
    for i in range(N):
        data = int(input(f"Enter data for node {i+1}: "))
        nodes[i] = Node(data)
        if i > 0:
            nodes[i-1].next = nodes[i]
    M = int(input("Enter number of random pointers: "))
    for i in range(M):
        src, dest = map(int, input(f"Enter source and destination indices for random pointer {i+1} separated by spaces: ").split())
        if 0 < src <= N and 0 < dest <= N:
            nodes[src-1].arb = nodes[dest-1]
    return nodes[0]

def test():
    head = createList()
    print("The original linked list is: ")
    printList(head)
    print()
    new_head = copyList(head)
    print("The copied linked list is: ")
    printList(new_head)

test()
```

```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def oddEvenList(head):
    if head is None or head.next is None:
        return head
    odd_head = head
    even_head = head.next
    odd_tail = odd_head
    even_tail = even_head
    curr = head.next.next
    is_odd = True
    while curr:
        if is_odd:
            odd_tail.next = curr
            odd_tail = curr
        else:
            even_tail.next = curr
            even_tail = curr
        curr = curr.next
        is_odd = not is_odd
    odd_tail.next = even_head
    even_tail.next = None
    return odd_head

def printList(node):
    while node:
        print(node.data, end=" ")
        node = node.next

def createList():
    n = int(input("Enter the number of nodes: "))
    nodes = [None] * n
    for i in range(n):
        data = int(input(f"Enter data for node {i+1}: "))
        nodes[i] = Node(data)
        if i > 0:
            nodes[i-1].next = nodes[i]
    return nodes[0]


def test():
    head = createList()
    print("The original linked list is: ")
    printList(head)
    print()
    new_head = oddEvenList(head)
    print("The reordered linked list is: ")
    printList(new_head)

test()
```

```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def nextLargerNodes(head):
    if head is None:
        return []
    values = []
    curr = head
    while curr:
        values.append(curr.data)
        curr = curr.next
    answer = [0] * len(values)
    stack = []
    for i in range(len(values)):
        while stack and values[i] > values[stack[-1]]:
            answer[stack.pop()] = values[i]
        stack.append(i)
    return answer

def printList(node):
    while node:
        print(node.data, end=" ")
        node = node.next

def createList():
    n = int(input("Enter the number of nodes: "))
    nodes = [None] * n
    for i in range(n):
        data = int(input(f"Enter data for node {i+1}: "))
        nodes[i] = Node(data)
        if i > 0:
            nodes[i-1].next = nodes[i]
    return nodes[0]

def test():
    head = createList()
    print("The original linked list is: ")
    printList(head)
    print()
    result = nextLargerNodes(head)
    print("The next greater node values are: ")
    print(result)

test()
```

```python

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def removeZeroSumSublists(head):
    if head is None:
        return None
    dummy = Node(0)
    dummy.next = head
    prefix_sums = {}
    curr_sum = 0
    curr = dummy
    while curr:
        curr_sum += curr.data
        if curr_sum in prefix_sums:
            start = prefix_sums[curr_sum]
            p = start.next
            sub_sum = curr_sum
            while p != curr:
                sub_sum += p.data
                p.data = 0
                del prefix_sums[sub_sum]
                p = p.next
            start.next = curr.next
        else:
            prefix_sums[curr_sum] = curr
        curr = curr.next
    return dummy.next

def printList(node):
    # Loop until the node is None
    while node:
        print(node.data, end=" ")
        node = node.next

def createList():
    n = int(input("Enter the number of nodes: "))
    nodes = [None] * n
    for i in range(n):
        data = int(input(f"Enter data for node {i+1}: "))
        nodes[i] = Node(data)
        if i > 0:
            nodes[i-1].next = nodes[i]
    return nodes[0]

def test():
    head = createList()
    print("The original linked list is: ")
    printList(head)
    print()
    new_head = removeZeroSumSublists(head)

    print("The new linked list is: ")
    printList(new_head)


test()
```
